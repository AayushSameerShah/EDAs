# 2. What is the Percentage of missing Females, Males and Children in top 10 regions of incidents?

```python
data.sample(3)
```

```python
missing = data[data['Region of Incident'].isin(top10.index) ].groupby(['Region of Incident'])[['Total Dead and Missing', 'Number of Females', 'Number of Males', 'Number of Children']].sum()
missing
```

### Based on the Number of missing people (NaN value affecting not sum to 100%)

```python
missing['females'] = (missing['Number of Females'] / missing['Total Dead and Missing']) * 100
missing['males'] = (missing['Number of Males'] / missing['Total Dead and Missing']) * 100
missing['children'] = (missing['Number of Children'] / missing['Total Dead and Missing']) * 100
missing.drop(missing.iloc[:, [0, 1, 2, 3]], axis= 1)
```

# 

### Based on their ratio

```python
missing_total = missing['Number of Females'] + missing['Number of Males'] + missing['Number of Children']

missing['females'] = (missing['Number of Females'] / missing_total) * 100
missing['males'] = (missing['Number of Males'] / missing_total) * 100
missing['children'] = (missing['Number of Children'] / missing_total) * 100
missing.drop(missing.iloc[:, [0, 1, 2, 3]], axis= 1)
```

# Q1

```python
ambiguous = data.groupby(["Name", "Gender"])['Count'].sum().copy()
ambiguous
```

```python
ambiguous = ambiguous.unstack()
ambiguous
```

```python
both_genders = ambiguous[(ambiguous.F.notna()) & (ambiguous.M.notna())].copy()
both_genders
```

### There are `1309` total names with both genders.

```python
both_G_names = both_genders.index
both_G_names
```

### Which year had the highest count of each

```python
data
```

#### A DF for only those names

```python
those_data = data[data.Name.isin(both_G_names)]
those_data
```

```python
those_data = those_data.groupby(['Name', 'Year'])['Count'].sum().unstack()
those_data
```

```python
max_year = those_data.idxmax(1)
max_year.name = "Year"

max_count = those_data.max(1)
max_count.name = "Count"
```

```python
temp = pd.concat([max_year, max_count], axis= 1)
temp
```

# Q3. Common names statewise
    Here we would have to give a threshold value. So that we can discriminate between common and uncommon

```python
data
```

#### Top 5 statewise

```python
statewise = data.groupby(['State', 'Name'])['Count'].sum()
statewise
```

# AMAZING!

```python
top5 = statewise.groupby(["State"]).apply(lambda x: x.nlargest(5))
top5
```

```python
top5 = top5.droplevel(0)
top5
```

# Q4 Name preferences timewise

```python
data
```

Again, the same thing... here we will choose TOP 3 .

```python
top3 = data.groupby(['Year', 'Name'])['Count'].sum()
top3 = top3.groupby(['Year']).apply(lambda x: x.nlargest(3)).droplevel(0)
top3
```
